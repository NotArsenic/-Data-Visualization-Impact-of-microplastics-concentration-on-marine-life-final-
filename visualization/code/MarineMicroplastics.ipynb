```python
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt

# Load shapefile or GeoJSON for world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Create a GeoDataFrame for microplastics detection data
# Replace 'your_data.csv' with the actual file path to your data
microplastics_data = gpd.GeoDataFrame.from_file('MarineMicroplastics.csv')

# Convert latitude and longitude columns to Point geometries
geometry = [Point(xy) for xy in zip(microplastics_data['Longitude'], microplastics_data['Latitude'])]

# Set the geometry of the GeoDataFrame to the Point geometries
microplastics_data = gpd.GeoDataFrame(microplastics_data, geometry=geometry)

# Plotting the world map
ax = world.plot(figsize=(10, 6), color='lightgrey', edgecolor='black')

# Plot microplastics detection data on top of the world map
microplastics_data.plot(ax=ax, color='red', markersize=5, alpha=0.5)

plt.title('World Map with Microplastics Detection')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sns
```

```python
#Histogram of Density Range of microplastics
df = pd.read_csv("MarineMicroplastics.csv")
plt.figure(figsize=(8, 6))
sns.histplot(df['Density Range'], bins=10, kde=True)
plt.title('Histogram of Density Range of Microplastics')
plt.xlabel('Density Range')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotate the x-axis labels by 45 degrees
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```

```python
# 3. Bar Chart of Density Class of microplastics
plt.figure(figsize=(8, 6))
sns.countplot(x='Density Class', data=df, order=df['Density Class'].value_counts().index)
plt.title('Density Class Distribution of Microplastics')
plt.xlabel('Density Class')
plt.ylabel('Count of Microplastics')
plt.show()
```

```python
# Load the dataset
df = pd.read_csv("MarineMicroplastics.csv")

# Create a bar plot showing density class frequencies for each density range
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Density Range', hue='Density Class', order=df['Density Range'].unique())
plt.title('Distribution of Microplastics Density Classes Across Density Ranges')
plt.xlabel('Density Range')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Density Class')
plt.tight_layout()
plt.show()
```

```python
# 4. Box Plot of Measurement by Oceans
plt.figure(figsize=(10, 6))
sns.boxplot(x='Oceans', y='Measurement', data=df)
plt.title('Measurement of Microplastics by Oceans')
plt.xlabel('Oceans')
plt.ylabel('Measurement in pieces/m3')
plt.xticks(rotation=45)
plt.show()
```

```python
#pie chart for showing different sampling methods for microplastics detection
import matplotlib.pyplot as plt
import pandas as pd
MarineMicroplastics = pd.read_csv('MarineMicroplastics.csv')
sampling_method_counts = MarineMicroplastics['Sampling Method'].value_counts()
N = 5  
top_methods = sampling_method_counts.nlargest(N)
plt.figure(figsize=(8, 8)) 
plt.pie(top_methods, labels=top_methods.index, autopct='%1.1f%%', startangle=140)
plt.title('Top Sampling Methods for Microplastics Detection')  
plt.axis('equal') 
plt.show()
```

```python
# 6. Scatter Plot of Measurement vs. Density Range
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Measurement', y='Density Range', hue='Density Class', data=df)
plt.title('Measurement vs. Density Range')
plt.xlabel('Measurement')
plt.ylabel('Density Range')
plt.legend(title='Density Class')
plt.show()
```

```python
# 7. Heatmap of Density Class by Ocean
heatmap_data = df.pivot_table(index='Oceans', columns='Density Class', aggfunc='size', fill_value=0)
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Density Class of Microplastics by Ocean')
plt.xlabel('Density Class')
plt.ylabel('Oceans')
plt.show()
```
